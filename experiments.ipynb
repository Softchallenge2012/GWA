```python
df_weighted = df.loc[df['dataset']=='weighted'].copy()
df_all = df.loc[df['dataset']=='all'].copy()
```

```python
models = ['lr score ','MLP score ','MLPM score ']
networks = ['random','homophily','heterophily']
clr = ['r','g','b']
scales = [0.2, 0.5, 0.7]

plt.figure(figsize=[18, 9])
plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=0.3, wspace=0.3)

row = 0
col = 0
for n in networks:

    for scale in scales:
        plt.subplot(331 + col)
        col +=1
        for i in range(3):
            plt.scatter(df_weighted.loc[df_weighted['model type']==models[i]].loc[df_weighted['scale']==scale].loc[df_weighted['network type']==n]['#samples'], \
                     df_weighted.loc[df_weighted['model type']==models[i]].loc[df_weighted['scale']==scale].loc[df_weighted['network type']==n]['accuracy'],c = clr[i])
            plt.title(f'{n} network N~(0, {scale})', loc='center')
            
            plt.ylim(0.4,1.2)
plt.legend(models)
```
